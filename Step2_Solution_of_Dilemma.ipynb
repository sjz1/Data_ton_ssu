# 전처리

```python
#필요없는 칼럼 제거
lib_in = lib_in.drop([lib_in.columns[0],lib_in.columns[1],lib_in.columns[5]], axis= 1)
lib_bor = lib_bor.drop([lib_bor.columns[0],lib_bor.columns[1],lib_bor.columns[2],lib_bor.columns[3],lib_bor.columns[4],lib_bor.columns[5],lib_bor.columns[9]], axis= 1)
# 중복되는 데이터 지우기
# 가정 (year_borrow admission_year,department) -> primary_key
primary_lib_in =lib_in.drop_duplicates(['연출입건수' ,'입학년도','소속'])
primary_lib_bor =lib_bor.drop_duplicates(['연대출권수' ,'입학년도','소속'])
```

# 입학년도 2010~2021 년만 살리고 Cut

```python
primary_lib_in =primary_lib_in[2021>=primary_lib_in.입학년도]
primary_lib_in =primary_lib_in[primary_lib_in.입학년도>=2010]
primary_lib_bor =primary_lib_bor[2021>=primary_lib_bor.입학년도]
primary_lib_bor =primary_lib_bor[primary_lib_bor.입학년도>=2010]
primary_lib_in.info()
```

#### [입학년도, 소속별] -> 연출입건수

```python
group_in = (primary_lib_in.groupby(["입학년도","소속"])["연출입건수"])
group_in.describe()
```

```python
mean_group_in = primary_lib_in.groupby(["입학년도","소속"])["연출입건수"].mean()
mean_group_in
```

```python
mean_group_in = pd.DataFrame(mean_group_in)
mean_group_in
```

#### [입학년도, 소속별] -> 연대출권수

```python
group_bor = (primary_lib_bor.groupby(["입학년도","소속"])["연대출권수"])
group_bor.describe()
```

```python
mean_group_bor = primary_lib_bor.groupby(["입학년도","소속"])["연대출권수"].mean()
mean_group_bor
```

```python
mean_group_bor = pd.DataFrame(mean_group_bor)
mean_group_bor
```

```python
join_group =pd.merge(mean_group_in,mean_group_bor, on = ['입학년도','소속'], how = 'outer')
join_group = join_group.dropna(how='any') #-> NA하나라도 있으면 지움
join_group.index = np.arange(len(join_group))
```

```python
join_group
```

```python
corr = join_group[['연출입건수','연대출권수']].corr(method = 'pearson')
print(corr)
```

```python
import seaborn as sns
cols_view = ['enter','borrow']
sns.set(font_scale = 1.5)
hm = sns.heatmap(corr.values,
                 cbar = True,
                 annot = True,
                 square = True,
                 fmt = '2f',
                 annot_kws = {'size' : 15},
                 xticklabels = cols_view,
                 yticklabels = cols_view
                )

plt.tight_layout()
plt.show
```

```python
x_val = []
y_val = []
for idx in range (0 ,len(join_group)):
    x_val.append(join_group["연출입건수"][idx])
    y_val.append(join_group["연대출권수"][idx])

plt.figure(figsize=(5, 5)) #그래프의 크기조절
#plt.xlim([0, 2000])      # X축의 범위: [xmin, xmax]
#plt.ylim([0, 100])     # Y축의 범위: [ymin, ymax]
plt.scatter(x_val,y_val,c='b',alpha= 0.5) #->그래프의 객체 생성
#alpha는 점들의 크기 label은 설명란
#plt.legend(loc = 'upper right') #밑에 Scatter point 박스의 위치를 선정 
plt.xlabel("연출입건수") #--->x축의 제목
plt.ylabel("연대출권수") #--->y축의 제목
plt.title("연출입건수 대비 연대출권수") # ---> 그래프 제목
plt.show()  #-> 그래프를 출력
```
