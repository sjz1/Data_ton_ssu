# 아쉬운대로 clustering 진행

```python
make_clu = []
for _ in range (0,len(join_group)):
    make_clu.append([])
    
for idx in range (0,len(join_group)):
    if join_group["연출입건수"][idx] < 1000 and join_group["연대출권수"][idx] < 250:
        make_clu[idx].append(join_group["연출입건수"][idx])
        make_clu[idx].append(join_group["연대출권수"][idx])
```

```python
while [] in make_clu:    
    make_clu.remove([])
make_clu
```

```python
from sklearn.cluster import KMeans
k= 4 # 클러스터의 개수를 미리 지정
kmeans = KMeans(n_clusters = k)
y_pred = kmeans.fit_predict(make_clu)
print("y_pred의 개수 : ",len(y_pred))
print(y_pred[:10])
```

```python
index_0 = []
index_1 = []
index_2 = []
index_3 = []

for idx in range (0,len(make_clu)):
    if y_pred[idx] == 0:
        index_0.append(make_clu[idx])
    elif y_pred[idx] == 1:
        index_1.append(make_clu[idx])
    elif y_pred[idx] == 2:
        index_2.append(make_clu[idx])
    else:
        index_3.append(make_clu[idx])
#len(index_0)+len(index_3)+len(index_2)+len(index_1)
```
