# 데이터 합병 진행 : bor_book = lib_bor + book

```python
bor_book =pd.merge(lib_bor,book, on = ['ISBN','서명'], how = 'outer')
bor_book.info()
```

```python
bor_book = bor_book.dropna(how='any') #-> NA하나라도 있으면 지움
bor_book.index = np.arange(len(bor_book))
bor_book.info()
```

```python
bor_book.head()
```

가정 : 같은 학과 같은 학번인 학우들 중 연대출권수가 같은 경우는 없다

```python
bor_book['연대출권수'] = bor_book['연대출권수'].astype(int)
bor_book['입학년도'] = bor_book['입학년도'].astype(int)
bor_book =bor_book.drop_duplicates(['연대출권수' ,'입학년도','소속'])
bor_book.info()
```

# 연대출권수와 입학년도의 상관 관계 확인

```python
#원본 유지
corr_bor_book= bor_book.drop([bor_book.columns[0],bor_book.columns[1],bor_book.columns[4],bor_book.columns[5]], axis= 1)
```

```python
corr_bor_book.head()
```

```python
tmp_group = pd.DataFrame(corr_bor_book.groupby(['입학년도'])["연대출권수"].mean())
tmp_group.head()
```

```python
lst_x = list(tmp_group.index)
lst_y = list(tmp_group['연대출권수'])
```

```python
Data_set = list(zip(lst_x ,lst_y))
df = pd.DataFrame(data = Data_set, columns=["입학년도","연대출권수"])
```

```python
df.head()
```

```python
corr = df[['연대출권수','입학년도']].corr(method = 'pearson')
print(corr)
```

```python
cols_view = ['borrow','enter_year']
sns.set(font_scale = 1.5)
hm = sns.heatmap(corr.values,
                 cbar = True,
                 annot = True,
                 square = True,
                 fmt = '2f',
                 annot_kws = {'size' : 15},
                 xticklabels = cols_view,
                 yticklabels = cols_view
                )

plt.tight_layout()
plt.show
```

```python
import matplotlib

matplotlib.rcParams['font.family'] ='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] =False

plt.figure(figsize=(5, 5)) #그래프의 크기조절
plt.title("입학년도와 연대출권수 관계")
plt.xlabel("입학년도")
plt.ylabel("연대출권수")
plt.bar(lst_x,lst_y, align = 'center')
```
